## How to use set parameters of general equation

As the required equations in Task 2 are similar, a general form of the equation is used in the following code

$$ \dot x = \alpha - ax^2 - b $$

where the parameters $\alpha$, a and b are handed over to the bifurcation-Function directly and to the equations_parameters of the phase_portrait-Function in a tuple

```python
bifurcation(alpha, 1, 0, name="equation_1")
phase_portrait(equation_parameters = (0, 1, 0), name="equation_1")
```

```python
bifurcation(alpha, 2, 3, name="equation_2")
phase_portrait(equation_parameters = (0, 2, 3), name="equation_2")
```
