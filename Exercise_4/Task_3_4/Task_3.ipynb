```python
from vae import *
```

```python
x_train,x_test,y_train,y_test = loadData()
```

# Training VAE - 2 dimensional Latent Space

```python
vae = VAE(encoder(latent_space=2), decoder(latent_space=2))
vae.compile(optimizer=keras.optimizers.Adam(learning_rate = 0.001))
latent_space=2
```

# 1st Epoch

```python
vae.fit(x_train, epochs=1, batch_size=128)
plotAll(vae, x_test, y_test,latent_space)
```

# 5th Epoch

```python
vae.fit(x_train, epochs=4, batch_size=128)
plotAll(vae, x_test, y_test,latent_space)
```

# 25th Epoch

```python
vae.fit(x_train, epochs=20, batch_size=128)
plotAll(vae, x_test, y_test,latent_space)
```

# 50th Epoch

```python
vae.fit(x_train, epochs=25, batch_size=128)
plotAll(vae, x_test, y_test,latent_space)
```

# 100th Epoch

```python
vae.fit(x_train, epochs=50, batch_size=128)
plotAll(vae, x_test, y_test,latent_space)
```
